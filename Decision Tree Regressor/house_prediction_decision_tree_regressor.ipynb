```python
import pandas as pd
path="train.csv"
df=pd.read_csv(path)
path1="test.csv"
df1=pd.read_csv(path1)
```

```python

from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor 
y=df.SalePrice
train_predictors= df.drop(['SalePrice'],axis=1)
X= train_predictors.select_dtypes(exclude=['object'])
X_train,X_valid,y_train,y_valid=train_test_split(X,y,train_size=0.8,test_size=0.2,random_state=0)

```

```python
def mae(max_nodes,train_X,val_X,train_y,val_y):
    model=DecisionTreeRegressor(max_leaf_nodes=max_nodes,random_state=0)
    model.fit(train_X,train_y)
    pred_val=model.predict(val_X)
    mae=mean_absolute_error(val_y,pred_val)
    return mae

```

```python
max_nodes=[5,25,50,100,150,250,500]
best=float('inf')
ans=0
for nodes in max_nodes:
    if best>mae(nodes,X_train,X_valid,y_train,y_valid):
        best=mae(nodes,X_train,X_valid,y_train,y_valid)
        ans=nodes

print(ans)
```

```python
test_X=df1.select_dtypes(exclude=['object'])
print(X.columns)
print(test_X.columns)
```

```python
final_model=DecisionTreeRegressor(max_leaf_nodes=ans,random_state=0)
final_model.fit(X,y)

pred_val=final_model.predict(test_X)
output=pd.DataFrame({'Id':df1.Id,'Sale Price':pred_val})
output.to_csv('results.csv',index=False)
```
